```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 最简单的plot绘图方式：传入y值
plt.plot()
plt.show()
```

```python
# 也可以同时传入x和y值
plt.plot([2, 4, 6, 8], [1, 4, 9, 16])
plt.show()
```

```python
# 另一种更加方便的传值方式，可以用于DataFrame、字典对象
data = pd.DataFrame({
    'round': [1, 2, 3, 4],
    'score': [4, 4.3, 3.8, 4.1]
})
plt.plot('round', 'score', data=data)
plt.show()
```

```python
# 如果传入多个序列，可以画出多条线
x = np.arange(1, 11)
plt.plot(x, x, x, np.sin(x), x, np.log(x))
plt.legend(labels=['linear', 'sine', 'logarithm'])
plt.show()
```

```python
# 也可以分多次绘制，方便对应数据与标签
x = np.arange(1, 11)
plt.plot(x, x, label='linear')
plt.plot(x, np.sin(x), label='sine')
plt.plot(x, np.log(x), label='logarithm')
plt.legend()
plt.show()
```
