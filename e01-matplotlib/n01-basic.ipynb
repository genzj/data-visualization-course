```python
%matplotlib inline
# 设置jupyter notebook的图形显示，普通Python脚本不需要

import matplotlib.pyplot as plt
# 导入pyplot，plt是惯用别名

import numpy as np
```

```python
x = np.arange(5)
y = [20,10,30,25,15]

# 此句可以注释掉
#plt.figure()


plt.bar(x,y, color='green', width=0.5)
plt.show()
```

```python
x = np.arange(5)
y = [20,10,30,25,15]
plt.figure()
plt.bar(x,y, color='green', width=0.5)

# 显式创建新图表
# 试试看注释了会怎样
plt.figure()

plt.scatter(x,y)
plt.show()
```
