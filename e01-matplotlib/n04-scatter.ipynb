```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
# 最简单的散点图只要传入等长的x和y轴数据即可
plt.scatter(tips['total_bill'], tips['tip'])
# 也可以用data方式
# plt.scatter('total_bill', 'tip', data=tips)
plt.xlabel('Total Bill $')
plt.ylabel('Tip $')
plt.show()
```

```python
# 定制散点图样式常使用如下几个参数
# c - 颜色，注意在RGB或RGBA格式时，使用两层嵌套列表
# s - 点大小
# marker - 点样式
# 完整列表：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter

plt.scatter(tips['total_bill'], tips['tip'], c=[[0.1, 0.2, 0.3, 0.5]], marker='^', s=100, edgecolors='k', linewidths=0.4)
plt.xlabel('Total Bill $')
plt.ylabel('Tip $')
plt.show()
```

```python
# 可以观察到，颜色、点大小等参数说明里都用复数，散点图各个参数均可以传输一个序列，作为高维度数据

sex_group = tips['sex'].map({'Male': 0, 'Female': 1})

plt.scatter(
    tips['total_bill'], tips['tip'],
    c=sex_group, alpha=0.6, # 颜色作为第三维度，此时可以统一设置透明度
    s=tips['size'] * 70, # 尺寸作为第四维度
    marker='^'
)
plt.xlabel('Total Bill $')
plt.ylabel('Tip $')
plt.show()
```

```python
# 思考：如果要用形状而不是颜色区分男女，该如何实现呢？
by_sex = tips.groupby('sex')
plt.scatter(
    'total_bill', 'tip',
    data=by_sex.get_group('Male'),
    s=100, marker='^', alpha=0.6,
    label='Male'
)
plt.scatter(
    'total_bill', 'tip',
    data=by_sex.get_group('Female'),
    s=100, marker='+', alpha=0.6,
    label='Female'
)
plt.legend()
plt.xlabel('Total Bill $')
plt.ylabel('Tip $')
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N) * 12  # 颜色也可以是连续值，mpl会使用cmap把各种值映射为具体的颜色
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)

# 此时一般会设置一个colorbar展示颜色和数值对应关系
color_bar = plt.colorbar()
color_bar.set_label('Area')

plt.show()
```

```python
sex_map = {'Male': 0, 'Female': 1}
sex_group = tips['sex'].map(sex_map)

plt.scatter(
    tips['total_bill'], tips['tip'],
    c=sex_group, alpha=0.6, # 颜色作为第三维度，此时可以统一设置透明度
    marker='^', s=90
)
plt.xlabel('Total Bill $')
plt.ylabel('Tip $')
plt.legend()

# 离散数据的colorbar比较麻烦，要设置好对应关系
colorbar = plt.colorbar()
colorbar.set_label('Sex')
sex_labels = np.array(list(sex_map.items()))
colorbar.set_ticks(sex_labels[:,1].astype('i'))    # 即把tick设置为0和1
colorbar.set_ticklabels(sex_labels[:,0]) # label设为对应的名称
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N) *12  # 颜色也可以是连续值，mpl会使用cmap把各种值映射为具体的颜色
area = (30 * np.random.rand(N))**2

# 使用cmap选择色彩映射， 完整列表参考 https://matplotlib.org/tutorials/colors/colormaps.html
for cmap in ['viridis', 'viridis_r', 'Greys', 'Set1']:
    plt.figure()
    plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap=cmap)
    plt.colorbar().set_label(cmap)
plt.show()
```
