````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
````

````python
# 设置符号为红点
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.show()
````

````python
# 所有符号一览
from matplotlib.lines import Line2D

points = np.ones(3)  # Draw 3 points for each line
text_style = dict(horizontalalignment='right', verticalalignment='center',
                  fontsize=12, fontdict={'family': 'monospace'})
marker_style = dict(linestyle=':', color='0.8', markersize=10,
                    mfc="C0", mec="C0")


def format_axes(ax):
    ax.margins(0.2)
    ax.set_axis_off()
    ax.invert_yaxis()

def nice_repr(text):
    return repr(text).lstrip('u').strip("'").strip('$')

def split_list(a_list):
    i_half = len(a_list) // 2
    return (a_list[:i_half] + ['$\\bigoplus$'], a_list[i_half:])

fig, axes = plt.subplots(ncols=2, figsize=(10, 12))

# Filter out filled markers and marker settings that do nothing.
markers = [m for m, func in Line2D.markers.items() if func != 'nothing' ]

for ax, markers in zip(axes, split_list(markers)):
    for y, marker in enumerate(markers):
        ax.text(-0.5, y, nice_repr(marker), **text_style)
        ax.plot(y * points, marker=marker, **marker_style)
        format_axes(ax)
plt.show()
````

````python
plt.figure(figsize=(5, 8))
plt.title(u'Color')
for y, color in enumerate(list('bgrcmykw') + ['pink', '#39de2b', (0, 0.5, 0.5), (0, 0, 0, 0.5)]):
    plt.text(-0.5, y, nice_repr(color), **text_style)
    lines = plt.plot(y * points, 'o', markerfacecolor=color, markeredgecolor='k', markersize=15)
    format_axes(lines[0].axes)

plt.show()
````

````python
# 用标记区分多个序列很方便
x = np.arange(1, 11)
plt.plot(x, x, 'ro', x, np.sin(x), 'g^', x, np.log(x), 'kx')
plt.show()
````

````python
plt.figure(figsize=(10, 4))
for y, line in enumerate(['-', '--', '-.', ':', (0, (10, 10)), (5, (10, 10)), (0, (15, 5, 2, 5))]):
    plt.text(-0.5, y, nice_repr(line), **text_style)
    lines = plt.plot(y * points, 'r', linestyle=line)
    format_axes(lines[0].axes)
plt.title('linestyle')
plt.show()
````

````python
# 线型可以一并设置
x = np.arange(1, 11)
plt.plot(x, x, 'ro-', x, np.sin(x), 'g^--')

# 复杂线型必须使用linestyle定义
plt.plot(x, np.log(x), 'kx', linestyle=(0, (10, 2, 1, 2)))
plt.show()
````

````python
# marker的颜色和大小也可以分开设置
x = np.arange(1, 11)
plt.plot(x, np.sin(x), '^--', markersize=15, markerfacecolor='r', markeredgecolor='b', markeredgewidth=1.5)
plt.show()
````

````python
# 标记可以间隔绘制，注意关键字是marksize，没有er
x = np.arange(0, 21) * np.pi / 8
plt.plot(x, np.sin(x), '^--', markevery=4, markersize=10)
plt.show()
````

````python
tips = sns.load_dataset('tips')
````

````python
plt.figure(figsize=(8, 5))
gr = tips.groupby('size')
tips_by_size = gr.sum()
plt.plot(tips_by_size.index, tips_by_size['total_bill'], 'r^-')
plt.plot(tips_by_size.index, tips_by_size['tip'], 'bo-')
plt.show()
````
