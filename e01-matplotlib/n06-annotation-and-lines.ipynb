```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

# annotate 核心是两个坐标——箭头端点坐标和文字坐标
plt.annotate(
    'local max',
    xy=(2, 1),
    xytext=(3, 1.5),
    arrowprops=dict(arrowstyle='->'),
)
plt.ylim(-2, 2)
plt.show()
```

```python
xycoords = [
    'figure points', 'figure pixels', 'figure fraction',
    'axes points', 'axes pixels', 'axes fraction', 'data', 'polar'
]
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

idx = 6
plt.annotate(
    xycoords[idx],
    xycoords=xycoords[idx],
    xy=(0.5, 0.5),
    xytext=(2.5, 1.5),
    textcoords='data',
    arrowprops=dict(arrowstyle='->'),
)
plt.ylim(-2, 2)
plt.show()
```

```python
# 文字位置可以相对定位
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate(
    xycoords[idx],
    xycoords=xycoords[idx],
    xy=(2, 1),
    xytext=(-100, 20),
    textcoords='offset pixels',
    arrowprops=dict(arrowstyle='->'),
)
plt.ylim(-2, 2)
plt.show()
```

```python
# 箭头相关属性可以通过arrowprops传入
# 可以使用预置的arrowstyle
styles = [
    '-', '->', '-[', '|-|', '-|>', '<-', '<->', '<|-', '<|-|>', 'fancy', 'simple', 'wedge'
]
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
plt.figure(figsize=(16, 10))
for idx, sty in enumerate(styles):
    plt.subplot(4, 3, idx + 1)
    line, = plt.plot(t, s, lw=2)
    plt.annotate(
        'style: ' + sty,
        xy=(2, 1),
        xytext=(3, 1.5),
        arrowprops=dict(arrowstyle=sty),
    )
    plt.ylim(-2, 2)
plt.show()
```

```python
# 进一步定制可以使用其他属性
# 完整的属性列表：https://matplotlib.org/api/_as_gen/matplotlib.pyplot.annotate.html#matplotlib.pyplot.annotate
# 和：https://matplotlib.org/api/_as_gen/matplotlib.patches.FancyArrowPatch.html#matplotlib.patches.FancyArrowPatch
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate(
    'local max',
    xy=(2, 1),
    xytext=(3, 1.7),
    arrowprops=dict(arrowstyle='->', edgecolor='b', linewidth=1.2, connectionstyle="arc3,rad=0.3")
)
plt.ylim(-2, 2)
plt.show()
```

```python
# 另有大量属性定义文字和文本框样式
# 参考：https://matplotlib.org/api/text_api.html#matplotlib.text.Text
# 如bbox给文字框加边线
# 参考：https://matplotlib.org/api/_as_gen/matplotlib.patches.FancyBboxPatch.html#matplotlib.patches.FancyBboxPatch
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate(
    'local max',
    xy=(2, 1),
    xytext=(3, 1.7),
    arrowprops=dict(arrowstyle='->', color='r'),
    fontsize='x-large',
    color='r',
    bbox=dict(boxstyle="round4", fc=(0.4, 0.9, 0.2, 0.4), linestyle=':')
)
plt.ylim(-2, 2)
plt.show()
```

```python
# 如果只要文字，无需箭头，直接用text函数即可，属性同annotate的Text部分
fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)


# 文字和Annotate的文字部分一样可以格式化及配框
ax.text(3, 8, 'boxed italics text in data coords', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})

ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)

ax.text(3, 2, 'unicode: Institut für Festkörperphysik')

ax.text(0.95, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='green', fontsize=15)

# 以上调用也可以基于pyplot语法在当前图形上调用
plt.text(0.5, 0.5, 'Middle',
        verticalalignment='center', horizontalalignment='center',
        transform=plt.gca().transAxes,
        color='red', fontsize=20)

# 在figure级别增加文字，可以使用figtext
plt.figtext(0, 0, '< figure bottom left')
plt.figtext(1, 1, '< figure upper right')

ax.axis([0, 10, 0, 10])

# 标题部位的文字有现成接口，可以使用pyplot设置或直接调用对象接口
# fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
# fig.subplots_adjust(top=0.85)
plt.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')
plt.subplots_adjust(top=0.85)

# 同理，以下两种写法是等价的，只是接口的层次不同
# ax.set_title('axes title')
# ax.set_xlabel('xlabel')
# ax.set_ylabel('ylabel')

plt.title('axes title')
plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.show()
```

```python
# 如果只要线，最简便的办法是增加水平或者垂直线
t = np.arange(-1, 2, .01)
s = np.sin(2 * np.pi * t)

plt.plot(t, s)
# 贯穿整个坐标的水平线
plt.axhline(y=0, linewidth=3, linestyle=':', color='#d62728')

# 同理可画垂直线
plt.axvline(x=1)

# 线可以有宽度，也可以有起止点，注意起止点是坐标轴的相对长度
plt.axvline(x=0, ymin=0.25, ymax=0.75, linewidth=8, color='k', alpha=0.5)
plt.axhline(y=.5, xmin=0.25, xmax=0.75, color='b')

plt.axis([-1, 2, -1, 2])

plt.show()
```

```python
# 如果只要线，最简便的办法是增加水平或者垂直线
t = np.arange(-1, 2, .01)
s = np.sin(2 * np.pi * t)

plt.plot(t, s)

# 以数据坐标画线，使用hlines和vlines，可以画单线或一组平行线
plt.hlines(np.arange(4) * 0.25, -0.5, 1.5, linewidth=1, linestyle=':', color='#d62728')
plt.vlines(np.arange(4) * 0.25, -0.5, 1.5, linewidth=1, linestyle=':', color='#d62728')


plt.axis([-1, 2, -1, 2])

plt.show()
```

```python
t = np.arange(-1, 2, .01)
s = np.sin(2 * np.pi * t)

plt.plot(t, s)

# 如果是高亮一个区间，可以使用axhspan或axvspan
# 区间有边又有面，可以组合多种样式
plt.axhspan(0.25, 0.75, facecolor='0.5', edgecolor='k', linewidth=2, alpha=0.5)

plt.axvspan(1.25, 1.55, facecolor='#2ca02c', edgecolor='r', linewidth=2, alpha=0.5)

plt.axis([-1, 2, -1, 2])

plt.show()
```
