```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 散点图同样常用且比matplotlib中更加方便
sns.scatterplot(
    'total_bill', 'tip', data=tips, hue='smoker'
)
plt.show()
```

```python
# matplotlib中Smoker一列的值是字符串，不能直接画图，需要转换
smoker_map = dict(Yes=1, No=0)
plt.scatter(
    'total_bill', 'tip', data=tips, c=tips.smoker.replace(smoker_map)
)
# 默认的legend也是错误的，需要手动定制
plt.legend()
plt.show()
```

```python
# 为了可读性考虑，散点图可以通过标志形状来区分
sns.scatterplot(
    'total_bill', 'tip', data=tips, style='smoker', s=111
)
plt.show()
```

```python
# 颜色、点的大小同样可以作为表现维度之一
sns.scatterplot(
    'total_bill', 'tip', data=tips, style='smoker', hue='sex', size=tips['size']
)
plt.show()
```

```python
# 当颜色是连续值时，会自动出现类似colorbar作用的图例
sns.scatterplot(
    'total_bill', 'tip', data=tips, style='smoker', hue=tips['size'], s=111
)
plt.show()
```

```python
# 大小也同样可以接收离散值
sns.scatterplot(
    'total_bill', 'tip', data=tips, size=tips['size']
)
plt.show()
```

```python
# 如果变化太多，不适合画在一个图上时，可以拆分开
plt.figure(figsize=(10, 6))
sns.scatterplot(
    'total_bill', 'tip', data=tips, hue='day', style='time'
)
plt.show()
```

```python
# 如果变化太多，不适合画在一个图上时，可以拆分开
sns.relplot(
    'total_bill', 'tip', data=tips, hue='day', row='day', col='time', s=111
)
plt.show()
```

```python
# 如果变化太多，不适合画在一个图上时，可以拆分开
sns.relplot(
    'total_bill', 'tip', data=tips, col='day', style='time', col_wrap=2, s=111
)
plt.show()
```
