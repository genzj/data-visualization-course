```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set()
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
# 绘制用餐人数和小费金额关系图，由于相同人数对应多种金额，于是出现了混乱
plt.plot('size', 'tip', 'b+-', data=tips)
plt.show()
```

```python
# 根据人数和小费金额进行排序，能得到相对整齐的图，但是可视化意义不大
plt.plot('size', 'tip', 'b+-', data=tips.sort_values(by=['size', 'tip']))
plt.show()
```

```python
# seaborn 能够自动聚合数据，并绘制出置信区间
sns.lineplot('size', 'tip', data=tips)
plt.plot()
```

```python
# 虽然原生绘图也可以绘制平均值，但是明显麻烦很多
by_size = tips.groupby('size').mean()
plt.plot(by_size.index, by_size.tip)
plt.show()
```

```python
# seaborn也向线性图引入了高维
# 比matplotlib更方便的是，其legend、颜色都是自动配置好的
sns.lineplot('size', 'tip', hue='sex', data=tips)
plt.show()
```

```python
# 要替换颜色也很简单
with sns.color_palette("Set2"):
    sns.lineplot('size', 'tip', hue='sex', data=tips)
plt.show()
# 完整色板参考：https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
# 线型也可作为更高维度
with sns.color_palette("Set2"):
    sns.lineplot('size', 'tip', hue='sex', style='smoker', markers='^o', markersize=10, data=tips)
plt.show()
```
